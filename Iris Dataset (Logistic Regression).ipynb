```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaled_x_train=scaler.fit_transform(x_train)
```

```python
scaled_x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
log_model=LogisticRegression(solver='saga',multi_class='ovr',max_iter=10000)
```

```python
penalty=['l1','l2','elasticnet']
l1_ratio=np.linspace(0,1,20)
C=np.logspace(0,10,20)
param_grid={'penalty':penalty,'l1_ratio':l1_ratio,'C':C}
```

```python
grid_model= GridSearchCV(log_model,param_grid=param_grid)
```

```python
grid_model.fit(scaled_x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report,plot_confusion_matrix,confusion_matrix
```

```python
grid_model.best_params_
```

```python
y_pred=grid_model.predict(scaled_x_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
grid_model.score(scaled_x_train,y_train)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import plot_roc_curve, roc_curve, auc
```

```python
def plot_multiclass_roc(clf, x_test, y_test, n_classes, figsize=(8,8)):
    y_score = clf.decision_function(x_test)

    # structures
    fpr = dict()
    tpr = dict()
    roc_auc = dict()

    # calculate dummies once
    y_test_dummies = pd.get_dummies(y_test, drop_first=False).values
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_dummies[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # roc for each class
    fig, ax = plt.subplots(figsize=figsize)
    ax.plot([0, 1], [0, 1], 'k--')
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title('Receiver operating characteristic example')
    for i in range(n_classes):
        ax.plot(fpr[i], tpr[i], label='ROC curve (area = %0.2f) for label %i' % (roc_auc[i], i))
    ax.legend(loc="best")
    ax.grid(alpha=.4)
    sns.despine()
    plt.show()
```

```python
plot_multiclass_roc(grid_model, scaled_x_test, y_test, n_classes =3)
```
